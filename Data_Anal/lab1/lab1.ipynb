```python
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, beta
import seaborn as sns
```

Exercise 1.1 & 1.2

```python


N = 1000


uniform_samples     = np.random.uniform(0, 1, N)
normal_std1_samples = np.random.normal(0, 1, N)
normal_std10_samples= np.random.normal(0, 10, N)
beta_11_samples     = np.random.beta(1, 1, N)
beta_22_samples     = np.random.beta(2, 2, N)
beta_39_samples     = np.random.beta(3, 9, N)
```

Exercise 1.3

```python

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
axes = axes.flatten()  


ax = axes[0]
ax.hist(uniform_samples, bins=25, density=True, alpha=0.5, label='Prbki')
x_u = np.linspace(0, 1, 200)
ax.plot(x_u, uniform.pdf(x_u, loc=0, scale=1), 'r', lw=2, label='PDF')
ax.set_title('Uniform(0,1)')
ax.legend()


ax = axes[1]
ax.hist(normal_std1_samples, bins=25, density=True, alpha=0.5, label='Prbki')
x_n1 = np.linspace(-4, 4, 200)
ax.plot(x_n1, norm.pdf(x_n1, loc=0, scale=1), 'r', lw=2, label='PDF')
ax.set_title('Normal(0,1)')
ax.legend()


ax = axes[2]
ax.hist(normal_std10_samples, bins=25, density=True, alpha=0.5, label='Prbki')
x_n10 = np.linspace(-50, 50, 300)  # Szerszy zakres ze wzgldu na due odchylenie
ax.plot(x_n10, norm.pdf(x_n10, loc=0, scale=10), 'r', lw=2, label='PDF')
ax.set_title('Normal(0,10)')
ax.legend()


ax = axes[3]
ax.hist(beta_11_samples, bins=25, density=True, alpha=0.5, label='Prbki')
x_b = np.linspace(0, 1, 200)
ax.plot(x_b, beta.pdf(x_b, 1, 1), 'r', lw=2, label='PDF')
ax.set_title('Beta(1,1)')
ax.legend()


ax = axes[4]
ax.hist(beta_22_samples, bins=25, density=True, alpha=0.5, label='Prbki')
ax.plot(x_b, beta.pdf(x_b, 2, 2), 'r', lw=2, label='PDF')
ax.set_title('Beta(2,2)')
ax.legend()


ax = axes[5]
ax.hist(beta_39_samples, bins=25, density=True, alpha=0.5, label='Prbki')
ax.plot(x_b, beta.pdf(x_b, 3, 9), 'r', lw=2, label='PDF')
ax.set_title('Beta(3,9)')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
mean_2d = [0, 0]


cov_1 = [[1, 0],
         [0, 1]]
cov_2 = [[2, 1],
         [1, 2]]


samples_cov1 = np.random.multivariate_normal(mean_2d, cov_1, size=N*10)
samples_cov2 = np.random.multivariate_normal(mean_2d, cov_2, size=N*10)

plt.figure(figsize=(12, 6))

# === Rozkad z cov_1 ===
plt.subplot(2,2,1)
plt.hist2d(
    samples_cov1[:, 0],
    samples_cov1[:, 1],
    bins=30,          
    density=True,     
    cmap='viridis'   
)
plt.colorbar(label='Gsto')
plt.title('hist2d: Cov = [[1,0],[0,1]],bins=30')

plt.subplot(2,2,2)
plt.hist2d(
    samples_cov1[:, 0],
    samples_cov1[:, 1],
    bins=70,          
    density=True,     
    cmap='viridis'    
)
plt.colorbar(label='Gsto')
plt.title('hist2d: Cov = [[1,0],[0,1]],bins=70')


plt.subplot(2,2,3)
plt.hist2d(
    samples_cov2[:, 0],
    samples_cov2[:, 1],
    bins=30,
    density=True,
    cmap='viridis'
)
plt.colorbar(label='Gsto')
plt.title('hist2d: Cov = [[2,1],[1,2]],bin=30')

plt.subplot(2,2,4)
plt.hist2d(
    samples_cov2[:, 0],
    samples_cov2[:, 1],
    bins=70,
    density=True,
    cmap='viridis'
)
plt.colorbar(label='Gsto')
plt.title('hist2d: Cov = [[2,1],[1,2]], bins=70')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10,5))


plt.subplot(1,2,1)
sns.kdeplot(
    x=samples_cov1[:, 0],
    y=samples_cov1[:, 1],
    fill=True,      
    cmap='viridis',
    cbar=True       
)
plt.title('KDE plot: Cov = [[1,0],[0,1]]')


plt.subplot(1,2,2)
sns.kdeplot(
    x=samples_cov2[:, 0],
    y=samples_cov2[:, 1],
    fill=True,
    cmap='viridis',
    cbar=True
)
plt.title('KDE plot: Cov = [[2,1],[1,2]]')

plt.tight_layout()
plt.show()
```

Exercise 1.5

```python
# Przygotuj tablice do przechowania wynikw
alpha_vals = np.empty(N)
beta_vals  = np.empty(N)
x_vals     = np.empty(N)
sigma_vals = np.empty(N)
y_vals     = np.empty(N)

for i in range(N):
    
    alpha = np.random.normal(loc=2.0, scale=0.1)
    beta  = np.random.normal(loc=0.2, scale=0.2)
    x     = np.random.normal(loc=0.0, scale=3.0)
    
    
    sigma = abs(np.random.normal(loc=0.0, scale=1.0))

    
    mu = alpha + beta * x

    
    y = np.random.normal(loc=mu, scale=sigma)

  
    alpha_vals[i] = alpha
    beta_vals[i]  = beta
    x_vals[i]     = x
    sigma_vals[i] = sigma
    y_vals[i]     = y
```

Exercise 1.6

```python
plt.figure(figsize=(8, 6))


plt.scatter(x_vals, y_vals, alpha=0.5, label='Wylosowane punkty (x, y)')


x_min, x_max = x_vals.min(), x_vals.max()
margin = 0.1 * (x_max - x_min)  
x_line = np.linspace(x_min - margin, x_max + margin, 300)


np.random.seed(123)
idxs = np.random.choice(len(alpha_vals), size=5, replace=False)


colors = ['red', 'green', 'orange', 'purple', 'black']

for j, i in enumerate(idxs):
    a_i = alpha_vals[i]
    b_i = beta_vals[i]
    y_line = a_i + b_i * x_line
    
    
    color = colors[j % len(colors)]
    plt.plot(x_line, y_line,
             linestyle='-', linewidth=2, alpha=0.8,
             color=color,
             label=f'Linia: alpha={a_i:.2f}, beta={b_i:.2f}')

plt.title('Porwnanie punktw (x,y) z liniami y = alpha + beta*x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
